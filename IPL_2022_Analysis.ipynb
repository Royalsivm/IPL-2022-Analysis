## The number of matches won by each team in IPL 2022

```python
sns.histplot(data=data, x="match_winner")
plt.xticks(rotation=90)
plt.figure(figsize=(16,8))
```

## Let's analyze whether most of the teams win by defending (batting first) or chasing (batting second).

```python
values = data["won_by"].value_counts(normalize=True)
plt.pie(values, labels=values.index, autopct="%1.1f%%")
plt.title("Match won by Wickets and Runs")
plt.show()
```

## What most teams prefer (batting or fielding) after winning the toss

```python
toss = data["toss_decision"].value_counts()
label = toss.index
counts = toss.values
colors = ['skyblue','yellow']
plt.figure(figsize=(4,4))
plt.pie(counts, labels=label, autopct="%1.1f%%", colors=colors)
plt.title("Toss Discision")
plt.show()
```

Thus, most captains choose to field after winning the toss.

## The top scorers of most IPL 2022 matches

```python
data["top_scorer"].value_counts()
x = data["top_scorer"].value_counts().index.to_list()
y = data["top_scorer"].value_counts().to_list()
plt.bar(x, y)
plt.xlabel("Top Scorers")
plt.ylabel("Count of Matches")
plt.title("The top scorers of most IPL 2022 match")
plt.xticks(rotation=90)
plt.figure(figsize=(4,70))
```

Jos Buttler has been a top scorer in 5 matches.

Lets analyze it deeply by including the runs scored by the top scorers.

```python
x = data["top_scorer"]
y = data["highscore"] 
plt.bar(x, y)
plt.xlabel("Top Batsman")
plt.ylabel("Runs scored")
plt.title("The top scorers of most IPL 2022 match")
plt.xticks(rotation=90)
plt.tight_layout()
plt.show()
```

 Jos Buttler has scored three centuries, and KL Rahul has scored two centuries.

## Who won the most " PLAYER OF THE MATCH " award in IPL 2022.

```python
data["player_of_the_match"].value_counts()
x = data["player_of_the_match"].value_counts().index.to_list()
y = data["player_of_the_match"].value_counts().to_list()
plt.bar(x, y)
plt.xlabel("Top Scorers")
plt.ylabel("Count of Matches")
plt.title("The top scorers of most IPL 2022 match")
plt.xticks(rotation=90,fontsize=8) 
#plt.figure(figsize=(46,13))

plt.tight_layout()

plt.show()
```

So Kuldeep Yadav is leading in the list of players of the match awards with four matches.

## The bowlers with the best bowling figures in most of the matches.

```python
data["best_bowling"].value_counts()
x = data["best_bowling"].value_counts().index.to_list()
y = data["best_bowling"].value_counts().to_list()
plt.bar(x, y)
plt.xlabel("Bowler")
plt.ylabel("Count of Matches")
plt.title("Best Bowlers in IPL 2022")
plt.xticks(rotation=90)
#plt.figure(figsize=(16,8))
plt.tight_layout()
plt.show()
```

 Yuzvendra Chahal having the best bowling figures in four matches.

## In which innings most of the wicket got taken.

In the Wankhede Stadium in Mumbai and MCA Stadium in Pune, most wickets fall while chasing the target. And in the other two stadiums, most wickets fall while setting the target.
